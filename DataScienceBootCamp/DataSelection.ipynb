```python
# import pandas and read the traffic file
import pandas as pd
df = pd.read_csv('DataSets/Traffic.csv')
# print the info
df.info()
```

```python
import matplotlib.pyplot as plt
selectSingleCol = df['Total']

# Plotting the line graph
plt.plot(selectSingleCol.index, selectSingleCol.values)
plt.title('Total Vehicles Over Time')
plt.xlabel('Entry Index')
plt.ylabel('Total Vehicles')

# Display the plot
plt.show()
```

```python
# select mulitable columns
multiableCol = df[['CarCount', 'BusCount', 'Day of the week', 'Total']]
multiableCol.sample(20)
```

```python
# slice a specific row by it's integer value
row = df.iloc[100]
print(row)
```

```python
# select multiple rows by their integer values
rows = df.iloc[100:150]
# it will start from 100 and end at 149
print(rows)
```

```python
# set an index for the dataframe
df_indexed = multiableCol.set_index('Day of the week')
print(df_indexed)
```

# Advanced indexing techniques
- Boolean indexing
- using query() method
- multi-level indexing

```python
# we only select rows that have a total number of vehicles > 1000
havvyTraffic = multiableCol[multiableCol['Total'] > 100]
print(havvyTraffic)
```

```python
# we can also do the same thing with query() method
havvyTraffic1 = multiableCol.query('Total >= 100')
print(havvyTraffic1)
```
