```python
import pandas as pd
# read the data from the traffic.csv file
df = pd.read_csv('/Users/tariq_assan/T5_test/DataSets/Traffic.csv')
# info() method to print the info of the dataframe
df.info()
```

```python
# print the first 5 rows of the dataframe
df.head()
```

```python
# print the last 5 rows of the dataframe
df.tail()
```

```python
# print a random sample of the dataframe
df.sample()
```

```python
# print the summary statistics of the dataframe
df.describe()
```

```python
# from the info() method we can see that three are 5 numerical columns in the dataframe.
"""
 0   Time               5952 non-null   object
 1   Date               5952 non-null   int64 
 2   Day of the week    5952 non-null   object
 3   CarCount           5952 non-null   int64 
 4   BikeCount          5952 non-null   int64 
 5   BusCount           5952 non-null   int64 
 6   TruckCount         5952 non-null   int64 
 7   Total              5952 non-null   int64 
 8   Traffic Situation  5952 non-null   object
"""
# so we are going to select the numerical columns and,
# define a relationship between them using corr() method.
numerical_columns = df[['CarCount','BikeCount','BusCount','TruckCount','Total']]

correlation_matrix = numerical_columns.corr()
print(correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# plot the heatmap of the correlation matrix
plt.figure(figsize=(10,6))
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of the Numerical Columns')
plt.show()
```

```python
# Pie chart with plotly express
import plotly.express as px
fig = px.pie(df, names='Traffic Situation', title='Traffic Situation Distribution')
fig.show()
```
